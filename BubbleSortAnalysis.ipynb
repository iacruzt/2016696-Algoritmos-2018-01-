**_Bubble sort analysis:_**

The following is a breaf analysis on bubble sort's performance. We'll assess three metrics: Number of instructions, comparisons and swaps. 

*Imports:*

```python
%matplotlib inline
import itertools # To generate the set of all permutations based on a given permutation. 
import matplotlib.pyplot as plotter # To paint the metrics.
import random # To (indirectly) generate a set of random permutations.
```

*Bubble sort assessment method:*

The following method provides a way to determine how many instructions, comparisons and swaps are needed to order a given permutation by the bubble sort algorithm. 

```python
def bubbleSortMetrics(permutation):
    numInstructions = 0
    numComparisons = 0
    numSwaps = 0
    for i in range(len(permutation)): # One instruction.
        numInstructions += 1
        for j in range(len(permutation) - i - 1): # One instruction.
            numInstructions += 4
            numComparisons += 1
            if permutation[j] > permutation[j + 1]: # Three instructions - One of which is a comparison.
                aux = permutation[j] # Two instructions.
                permutation[j] = permutation[j + 1] # Two instructions - One of which is a swap.
                permutation[j + 1] = aux # One instruction.
                numInstructions += 5
                numSwaps += 1
    return (numInstructions, numComparisons, numSwaps)
```

*Bubble sort assessment for all the permutations of the first permutationLength (a variable) natural numbers:*

[Notice, by the way, that given the nature of the algorithm it will take the same number of comparisons for any two permutations of the same length; ergo, all these permutations to test the algorithm's performance will register the same number of comparisons.] 

```python
permutationLength = 8
numsInstructions = []
numsComparisons = []
numsSwaps = []
for permutation in itertools.permutations(list(range(permutationLength))):
    numInstructions, numComparisons, numSwaps = bubbleSortMetrics(list(permutation))
    numsInstructions.append(numInstructions)
    numsComparisons.append(numComparisons)
    numsSwaps.append(numSwaps)
avgNumInstructions = sum(numsInstructions)/len(numsInstructions)
avgNumComparisons = sum(numsComparisons)/len(numsComparisons)
avgNumSwaps = sum(numsSwaps)/len(numsSwaps)
```

```python
plotter.hist(numsInstructions)
plotter.xlabel("# of instructions")
plotter.ylabel("# of permutations")
plotter.title("Freq. Distribution for the \n # of instructions")
plotter.show()
```

```python
plotter.hist(numsComparisons)
plotter.xlabel("# of comparisons")
plotter.ylabel("# of permutations")
plotter.title("Freq. Distribution for the \n # of comparisons")
plotter.show()
```

```python
plotter.hist(numsSwaps)
plotter.xlabel("# of swaps")
plotter.ylabel("# of permutations")
plotter.title("Freq. Distribution for the \n # of swaps")
plotter.show()
```

*Average values:*

```python
print("AVG number of instructions: {}".format(avgNumInstructions))
print("AVG number of comparisons: {}".format(avgNumComparisons))
print("AVG number of swaps: {}".format(avgNumSwaps))
```

*Method to compute factorial function:*

[Why the factorial function? Look at the next text box...]

```python
def factorial(n):
    if n < 1:
        return 1
    else:
        return factorial(n-1)*n
```

*Bubble sort assessment for (permutationLength)! permutations based on the first permutationLength^2 natural numbers, from which permutationLength elements are randomly picked:*

[Notice, by the way, that given the nature of the algorithm it will take the same number of comparisons for any two permutations of the same length; ergo, all these permutations to test the algorithm's performance will register the same number of comparisons.] 

```python
permutationLength = 8
numsInstructions = []
numsComparisons = []
numsSwaps = []
for permutation in [random.sample(list(range(permutationLength**2)), permutationLength) for i in range(factorial(permutationLength))]:
    numInstructions, numComparisons, numSwaps = bubbleSortMetrics(list(permutation))
    numsInstructions.append(numInstructions)
    numsComparisons.append(numComparisons)
    numsSwaps.append(numSwaps)
avgNumInstructions = sum(numsInstructions)/len(numsInstructions)
avgNumComparisons = sum(numsComparisons)/len(numsComparisons)
avgNumSwaps = sum(numsSwaps)/len(numsSwaps)
```

```python
plotter.subplot(3,1,1)
plotter.hist(numsInstructions)
plotter.xlabel("# of instructions")
plotter.ylabel("# of permutations")
plotter.title("Freq. Distribution for the \n # of instructions")
plotter.show()
```

```python
plotter.subplot(3,1,2)
plotter.hist(numsComparisons)
plotter.xlabel("# of comparisons")
plotter.ylabel("# of permutations")
plotter.title("Freq. Distribution for the \n # of comparisons")
plotter.show()
```

```python
plotter.subplot(3,1,3)
plotter.hist(numsSwaps)
plotter.xlabel("# of swaps")
plotter.ylabel("# of permutations")
plotter.title("Freq. Distribution for the \n # of swaps")
plotter.show()
```
